```python
import os, json, pandas as pd
from dateutil import tz
from dateutil import parser as dateparser
from datetime import datetime
```

```python
df = pd.read_csv('dataset/Camera_Traffic_Counts_20251108.csv')
df.head()
```

```python
# Clean column names: strip whitespace, convert to lowercase, replace spaces with underscores
df.columns = (df.columns
              .str.strip()
              .str.lower()
              .str.replace(r"\s+", "_", regex=True))

print("\n Clean Columns:")
print('-----------------------------------')
print(df.columns)

print("\n Data:")
print('-----------------------------------')
df.head()
 
```

```python
print("\nData Shape:")
print('-----------------------------------')
print(df.shape)

print("\nData Types:")
print('-----------------------------------')
print(df.dtypes)

print("\nInfo:")
print('-----------------------------------')
print(df.info())
```

```python
# convert "Month","Day","Year","Hour","Minute" into a single datetime column
# Ensure the source columns exist, convert to numeric (coerce errors), then build a datetime column
cols = ['year', 'month', 'day', 'hour', 'minute']
for c in cols:
    if c in df.columns:
        # convert to numeric; invalid parsing becomes NaN
        df[c] = pd.to_numeric(df[c], errors='coerce')
    else:
        print(f"Warning: column '{c}' not found in dataframe")

# Create datetime object
#df['datetime'] = datetime(year, month, day, hour, minute)
```

```python
print("\nData Types:")
print('-----------------------------------')
print(df.dtypes)
```

```python
# Check if any record has null values
df_has_null = df.isnull().any().any()

# List of columns with null values
columns_with_null = df.columns[df.isnull().any()].tolist()

print(df_has_null)           # True if any nulls exist
print(columns_with_null)     # Columns that contain null values
```
